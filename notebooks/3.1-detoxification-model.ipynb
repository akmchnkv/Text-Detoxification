```python
import pandas as pd
import torch

df = pd.read_csv('../data/interim/filtered_df.tsv', sep='\t')
```

```python
columns_to_drop = ['similarity', 'length_diff', 'ref_tox', 'trn_tox']
df.drop(columns=columns_to_drop, inplace=True)
df.head()
```

Let choice 10 random phrases from dataset

```python
random_phrases = df.sample(10)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

# Visualization of the results

```python
import matplotlib.pyplot as plt

def plot_toxicity(reference_toxic_scores, model_toxic_scores, model):
    plt.figure(figsize=(8, 6))
    plt.plot(reference_toxic_scores, marker='o', color='blue', label='Reference Toxicity')
    plt.plot(model_toxic_scores, marker='o', color='green', label= model + ' toxicity')
    plt.xlabel('Phrase Index')
    plt.ylabel('Toxicity Score')
    plt.title('Toxicity Comparison')
    plt.legend()
    plt.show()
```

# Bart model

```python
import torch
from transformers import BartForConditionalGeneration, BartTokenizer

bart_model = BartForConditionalGeneration.from_pretrained('SkolkovoInstitute/bart-base-detox')
tokenizer = BartTokenizer.from_pretrained('SkolkovoInstitute/bart-base-detox')
```

```python
pip install -e detoxify
```

```python
from detoxify import Detoxify

reference_toxic_scores = []
bart_toxic_scores = []
translation_toxic_scores = []

i = 0
for idx, row in random_phrases.iterrows():
    i = i + 1
    print('P H R A S E', i)
    
    reference_phrase = row['reference']
    translation_phrase = row['translation']
    
    input_ids = tokenizer.encode(reference_phrase, return_tensors='pt').to(device)

    bart_model.to(device)
    bart_model.eval()
    
    with torch.no_grad():
        input_ids = input_ids.to(device)
        generated_ids = bart_model.generate(input_ids, max_length=100, num_return_sequences=1)
        model_prediction = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)
    
    reference_toxicity = Detoxify('original').predict(reference_phrase)['toxicity']
    model_toxicity = Detoxify('original').predict(model_prediction[0])['toxicity']
    translation_toxicity = Detoxify('original').predict(translation_phrase)['toxicity']
    
    translation_toxic_scores.append(translation_toxicity)
    reference_toxic_scores.append(reference_toxicity)
    bart_toxic_scores.append(model_toxicity)

    print('Reference:', reference_phrase)
    print('Translation:', translation_phrase)
    print('Bart model translation:', model_prediction[0])
    print()
    print('Reference toxic score:', reference_toxicity)
    print('Bart model toxic score:', model_toxicity)
    print('-'*10)
```

```python
plot_toxicity(reference_toxic_scores, bart_toxic_scores, 'Bart')
```

# T5-small model

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

t5_model = T5ForConditionalGeneration.from_pretrained('t5-small')
tokenizer = T5Tokenizer.from_pretrained('t5-small')
```

```python
t5_toxic_scores = []

i = 0
for idx, row in random_phrases.iterrows():
    i = i + 1
    print('P H R A S E', i)
    
    reference_phrase = row['reference']
    translation_phrase = row['translation']
    
    input_ids = tokenizer.encode(reference_phrase, return_tensors='pt', padding=True, truncation=True).to(device)

    t5_model.to(device)
    t5_model.eval()
    
    with torch.no_grad():
        input_ids = input_ids.to(device)
        generated_ids = t5_model.generate(input_ids, max_length=100, num_return_sequences=1)
    model_prediction = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
    
    model_toxicity = Detoxify('original').predict(model_prediction)['toxicity']
    t5_toxic_scores.append(model_toxicity)

    print('Reference:', reference_phrase)
    print('Translation:', translation_phrase)
    print('T5 model translation:', model_prediction)
    print()
    print('Reference toxic score:', reference_toxic_scores[i-1])
    print('T5 model toxic score:', model_toxicity)
    print('-'*10)
```

```python
plot_toxicity(reference_toxic_scores, t5_toxic_scores, 'T5-small')
```
